```python
import json
import matplotlib.pyplot as plt
```

```python
f=open('grades.json')
b=json.load(f)
counter=1
plt.figure(figsize=(10,10))
for x in b.keys():
    plt.subplot(2,3,counter)
    counter +=1
    sum=0
    a=list(b[x].keys())
    a=a[::-1]
    for i in a:
        sum += b[x][i]
    prob=dict({})
    for i in a:
        prob[i] = b[x][i]/sum
    import matplotlib.pyplot as plt
    plt.plot(prob.keys(),prob.values(),"-o")
    mysum=0
    cdf=dict({})
    for i in a:
        mysum += prob[i]
        cdf[i]=mysum
    plt.step(cdf.keys(),cdf.values(),"-o") 
    plt.title(str(x))
    plt.plot()
```
